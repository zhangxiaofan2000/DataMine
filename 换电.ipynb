```python

def find_pos(start, end, l, count):
    '''
    查找最后一个正数,nlogn
    :param start: 0
    :param end: len(l)-1
    :param l: 列表
    :param count: 0
    :return: 返回index+1
    '''
    if(start > end):
        return count
    middle = start + (end-start)//2
    if(l[middle] >= 0):
        count += (middle-start) + 1
        return find_pos(middle+1, end, l, count)
    else:
        return find_pos(start, middle-1, l, count)

```

```python
def price(minute,start):
    '''
    计算该状态电费
    :param e:
    :param minute: 充电时长 分钟
    :param start: 开始充电的时间
    :return:
    '''
    start = start.hour*60+start.minute+start.second/60
    end_min = [ (start + minute) - i*60 for i in interval ]
    index = find_pos(0, len(end_min)-1, end_min, 0)-1
    #22点
    if index == 4:
        #超过24点
        if end_min[index] > 2*60:
            return electricity_fees[index] * (minute-end_min[index]+120)+ electricity_fees[0] * (end_min[index]-120)
        else:
            return electricity_fees[index] * minute

    if end_min[index]- minute >= 0:
        p = electricity_fees[index] * minute
    else:
        p = electricity_fees[index-1] * (minute- end_min[index]) + electricity_fees[index] * end_min[index]
    return p
```

```python
def find_next(x,df,index):
    '''
    寻找 仓位x 下一次出现的位置
    :param x: df['取车电池仓位']
    :return: 存车电池仓位 index
    '''
    index = index+1
    for i in range(index,len(df.index)):
        #13为
        if df.iloc[i,15] == x :
            return i
    return len(df.index)-1
```

```python


def Compare_fee(index,Earliest,Latest):
    '''
    比较以 Earliest 与 Latest 开始充电的费用
    :param index:
    :param Earliest:
    :param Latest:
    :return:
    '''
    if price(df_pro.iloc[index, 2], Earliest)<=price(df_pro.iloc[index, 2], Latest):
        return Earliest,price(df_pro.iloc[index, 2], Earliest)
    else:
        return Latest,price(df_pro.iloc[index, 2], Latest)


def Get_interval(start,end):
    '''
    获取时间是否在间隔interval_series内
    只需要用到时间的小时，分钟，秒
    :param start:
    :param end:
    :return:
    '''
    left = start.hour + start.minute / 60 + start.second / 3600
    right = end.hour + end.minute / 60 + end.second / 3600
    return  interval_series.between(left, right)

def Get_fee(Earliest,is_between,index):
    '''
    if sum(is_between) > 0 才可以用此函数计算不同情况下的费用
    :param is_between:
    :return:
    '''

    true_index = [i for i, val in enumerate(is_between.tolist()) if val]
    #获得时间
    true_index_time = [ pd.Timestamp(year=Earliest.year, month=Earliest.month, day=Earliest.day, hour=interval[i]) for i in true_index ]
    #获取各种情况下的费用
    return [price(df_pro.iloc[index, 2], time) for time in true_index_time],true_index_time

def Get_time(is_between,index):
    '''
    得到最优开始时间和最小费用
    :param is_between:
    :param Earliest:
    :param index:
    :return:
    '''
        #如果时间有重叠（详见pandas.series.between）
    if sum(is_between) > 0:
        fee,true_index_time = Get_fee(Earliest,is_between,index)
        #选取费用最小的开始充电时间
        if min(fee) < price(df_pro.iloc[index, 2], Earliest):
            min_fee = min(fee)
            ans =  true_index_time[fee.index(min(fee))]
        else:
            ans,min_fee= Compare_fee(index,Earliest,Latest)
    #时间无重叠
    else:
        ans,min_fee= Compare_fee(index,Earliest,Latest)

    return ans,min_fee

def opt(index,Earliest,Latest):
    '''
    给定开始与结束时间，搜索最优开始充电时间
    :param index:
    :param Earliest:
    :param Latest:
    :return:
    '''
    #同一天
    if Earliest.day == Latest.day:
        #获取间隔
        is_between = Get_interval(Earliest,Latest)

        ans,min_fee =  Get_time(is_between,index)


    #不是同一天
    else:
        split_day_1 = pd.Timestamp(year=Earliest.year, month=Earliest.month, day=Earliest.day, hour=23)
        split_day_2 = pd.Timestamp(year=Earliest.year, month=Earliest.month, day=Earliest.day, hour=0)
        is_between_1 =  Get_interval(Earliest,split_day_1)
        is_between_2 =  Get_interval(split_day_2,Latest)
        ans1,min_fee_1 =  Get_time(is_between_1,index)
        ans2,min_fee_2 =  Get_time(is_between_2,index)
        if min_fee_1 < min_fee_2:
            ans = ans1
            min_fee = min_fee_1
        else:
            ans= ans2
            min_fee = min_fee_2

    return ans,min_fee
```

```python
interval = [0, 8, 12, 18, 22]
interval_series = pd.Series(interval)
electricity_fees = [0.3484,1.0323,0.668,1.0323,0.668]

buffer_time=[]
start_time=[]
end_time=[]
all_fees=[]
original_fees=[]
for i,j in enumerate(Position):
    # 最早充电时间 取车电池结束时间
    Earliest = df_pro.iloc[i,10]

    # 最晚开始时间 该仓位下一次存车开始时间-充电时间
    Latest = df_pro.iloc[find_next(df_pro.iloc[i,13],df_pro,i),10] - pd.to_timedelta(df_pro.iloc[i,2], unit='m')
    # 计算优化后的时间费用
    ans,min_fee = opt(i,Earliest,Latest)

    #计算原始费用
    original_cost = price(df_pro.iloc[i, 2], Earliest)

    start_time.append(ans)
    end_time.append(ans+pd.to_timedelta(df_pro.iloc[i,2], unit='m'))

    all_fees.append(min_fee)
    original_fees.append(original_cost)

    #计算可优化空间
    buffer_time.append(Latest-Earliest)

df_pro['可搜索时间'] = buffer_time
df_pro['优化后费用'] = all_fees
df_pro['原始费用'] = original_fees
df_pro['优化后开始充电时间'] = start_time
df_pro['优化后结束充电时间'] = end_time
df_pro.loc[(df_pro['优化后开始充电时间']-df_pro['开始时间'])<pd.to_timedelta(0, unit='d'),'优化后开始充电时间'] = df_pro['优化后开始充电时间']+pd.to_timedelta(1, unit='d')

df_pro['优化后开始充电时间'] = pd.to_datetime(df_pro['优化后开始充电时间'],format="%Y-%m-%d %h:%m:%s",unit='s')
```
